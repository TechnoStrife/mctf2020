```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flag = open('flag.txt', 'r').readlines()
flag.pop(0)
flag = [list(map(float, x.strip().split(','))) for x in flag]
```

```python
train = open('train.txt', 'r').readlines()
train.pop(0)
def qwe(x):
    x = x.strip().split(',')
    x.append(x.pop(0))
    x[0] = float(x[0])
    x[1] = float(x[1])
    x.append(x[2])
    x[2] = 'o'+ {'0':'r', '1':'b'}[x[2]]
    return x
train = [qwe(x) for x in train]
```

```python
data = []
for x in train:
    data.extend(x[:3])
plt.plot(*data)
plt.plot(*zip(*flag), 'og')
None
```

```python
from math import sqrt

def dist(a, b):
    return sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)

def split(arr, l):
    return [arr[i:i+l] for i in range(0, len(arr), l)]

res = []
for q in flag:
    r = min(train, key = lambda tr: dist(q, tr))
    res.append(r[3])

print(''.join([chr(int(''.join(x), 2)) for x in split(res, 8)]))
```
